```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
#your code here
```

```python
spaceship = spaceship.dropna()
```

```python
spaceship["Cabin"] = spaceship["Cabin"].str.split("/").str[0]
```

```python
spaceship = spaceship.drop(columns = ["PassengerId" , "Name"])
```

```python
spaceship_d = pd.get_dummies(spaceship, columns=['HomePlanet', "Cabin", "Destination"], drop_first=True)
```

```python
spaceship_d.head(5)
```

**Perform Train Test Split**

```python
#your code here
```

```python
features = spaceship_d.drop(columns = "Transported")
```

```python
target = spaceship_d["Transported"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

normalizer = MinMaxScaler()

normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

**Grid/Random Search**

- Now let's use the best model we got so far in order to see how it can improve when we fine tune it's hyperparameters.

- Run Grid Search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
grid = {"n_estimators": [50, 100],
        "max_depth":[10,20]}

#ValueError: Invalid parameter 'estimator' for estimator RandomForestClassifier(). 
#Valid parameters are: ['bootstrap', 'ccp_alpha', 'class_weight', 'criterion', 
# 'max_depth', 'max_features', 'max_leaf_nodes', 'max_samples', 'min_impurity_decrease', 
#'min_samples_leaf', 'min_samples_split', 'min_weight_fraction_leaf', 'n_estimators', 
#'n_jobs', 'oob_score', 'random_state', 'verbose', 'warm_start']
```

```python
forest_h = RandomForestClassifier()
```

```python
model = GridSearchCV(estimator =forest_h , param_grid = grid, cv=5)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
best_model = model.best_estimator_
```

- Evaluate your model

```python
forest_h = RandomForestClassifier(n_estimators = 50,
                             max_depth = 10)
```

```python
forest_h.fit(X_train, y_train)
```

```python
forest_h.score(X_test, y_test)
```

```python
pred = best_model.predict(X_test)

pred = pred.astype(int) 
y_test = y_test.astype(int) 

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", best_model.score(X_test, y_test))
```
